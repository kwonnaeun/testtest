# 2. CRF 학습하여 Person 이름 제거

```python
#CRF 학습 준비를 위한 pos태깅
from nltk import pos_tag, word_tokenize
text = []
postag = []

for sent in news:
    text += sent.split('.')
    
for sent in text:
    postag.extend(pos_tag(word_tokenize(sent)))
```

```python
# 기사에 사용된 단어 POS 태깅
pos = pd.DataFrame(postag)
pos.columns=['Words','POS']
pos
```

```python
# IOB태깅 된 kaggle 데이터를 활용하여 데이터 학습
# Kaggle: https://www.kaggle.com/abhinavwalia95/entity-annotated-corpus/version/4

data = pd.read_csv("C:/Users/user/Desktop/과제/텍스트마이닝/ner_dataset.csv", encoding="latin1")
```

```python
data.columns = ['Sentence #', 'Word', 'POS', 'Tag']
```

```python
data = data.fillna(method="ffill")
```

```python
# "Word" 열의 값들 중 unique한 것들로 list 생성 
words = list(set(data["Word"].values))
n_words = len(words)

n_words
```

```python
# 데이터에서 문장을 원하는 형태로 포매팅하여 추출
class SentenceGetter(object):
    
    def __init__(self, data):
        self.n_sent = 1
        self.data = data
        self.empty = False
        agg_func = lambda s: [(w, p, t) for w, p, t in zip(s["Word"].values.tolist(),
                                                            s["POS"].values.tolist(),
                                                            s["Tag"].values.tolist())]
        self.grouped = self.data.groupby("Sentence #").apply(agg_func)
        self.sentences = [s for s in self.grouped]
```

```python
# 전체 문장 추출
getter = SentenceGetter(data)
sentences = getter.sentences
```

```python
sentences[0]
```

```python
#단어 단위 함수 정의
def word2features(sent, i):
    word = sent[i][0]
    postag = sent[i][1]

    features = {
        'word.lower()': word.lower(),
        'word[-3:]': word[-3:],
        'word[-2:]': word[-2:],
        'word.isupper()': word.isupper(),
        'word.istitle()': word.istitle(),
        'word.isdigit()': word.isdigit(),
        'postag': postag,
        'postag[:2]': postag[:2],
    }
    if i > 0:
        word1 = sent[i-1][0]
        postag1 = sent[i-1][1]
        features.update({
            '-1:word.lower()': word1.lower(),
            '-1:word.istitle()': word1.istitle(),
            '-1:word.isupper()': word1.isupper(),
            '-1:postag': postag1,
            '-1:postag[:2]': postag1[:2],
        })
    else:
        features['BOS'] = True

    if i < len(sent)-1:
        word1 = sent[i+1][0]
        postag1 = sent[i+1][1]
        features.update({
            '+1:word.lower()': word1.lower(),
            '+1:word.istitle()': word1.istitle(),
            '+1:word.isupper()': word1.isupper(),
            '+1:postag': postag1,
            '+1:postag[:2]': postag1[:2],
        })
    else:
        features['BOS'] = True

    return features

#문장 단위 함수 정의
def sent2features(sent):
    return [word2features(sent, i) for i in range(len(sent))]

#label만 저장
def sent2labels(sent):
    return [label for token, postag, label in sent]

#token만 저장
def sent2tokens(sent):
    return [token for token, postag, label in sent]
```

```python
X = [sent2features(s) for s in sentences]
y = [sent2labels(s) for s in sentences]
```

```python
from sklearn_crfsuite import CRF

crf = CRF(algorithm='lbfgs',
          c1=0.1, 
          c2=0.1,
          max_iterations=100, 
          all_possible_transitions=False)
```

```python
# Training
crf.fit(X[:], y[:])
```

```python
# POS 태깅한 기사 내용을 가지고 입력데이터 생성
X_test = [sent2features(postag)]
```

```python
y_test_pred = crf.predict(X_test)
```

```python
y_test_pred[0]
```

```python
print(len(postag))
print(len(y_test_pred[0]))
```

```python
# 예측한 태그 정보를 단어 데이터 프레임에 추가
pos['tag'] = y_test_pred[0]
```

```python
pos
```

```python
# Person 태그 단어 저장
tag_list = pos.values.tolist()
PER = []

for w, y, z in tag_list:
    if z=='B-per' or z=='I-per':
        PER.append(w.lower())
```

```python
PER
```

```python
#사전 리스트에서 Person 태그에 해당하는 단어제거 (영어 학습을 위한 사전 리스트에서 무의미한 단어)
count = 0
for i in y_test_pred[0]:
    if 'B-per' in i:        
        count+=1
    elif 'I-per' in i:
        count+=1
        
print(count)
print(len(postag))
print(count/len(postag))
```

# 3. 데이터 전처리

```python
import re
import nltk
import string
from nltk.corpus import stopwords  
from nltk.stem import PorterStemmer
import math
import pandas as pd
from pandas import DataFrame as df
import seaborn as sns
from nltk.corpus import stopwords
```

```python
# 숫자, 특수 문자 제거 및 불용어 제거
totalword =[]
train_word = []

for category_news in news:
    lowertext = [i.lower() for i in category_news]
    lowertext1 = ''.join(i for i in lowertext if not i.isdigit())
    lowertext2 = re.sub('\W+|[0-9]+', ' ', lowertext1)
    tokenedtext = lowertext2.split()
    tokenedtext = [x for x in tokenedtext if x not in PER] #Person 제거
    tokenedtext = [x for x in tokenedtext if x not in string.punctuation]
    tokenedtext = [x for x in tokenedtext if x not in nltk.corpus.stopwords.words('english')]
    train_word.append(tokenedtext)
    for word in tokenedtext:
        totalword.append(word)
```

```python
totalword
```

# 4. 빈도수 순으로 단어 사전 만들기 (Level1~5)

```python
freqdist = nltk.FreqDist(totalword)
freqdist = freqdist.most_common()
```

```python
freqdist
```

```python
word_dic = [w for (w,z) in freqdist] 
```

```python
word_dic
```

```python
len(word_dic)
```

```python
count = 0
split = math.ceil(len(word_dic) / 5)
new_dic = {}
for x in word_dic:
    count += 1
    if count < split:
        new_dic[x] = 1
    elif count < (split*2):
        new_dic[x] = 2
    elif count < (split*3):
        new_dic[x] = 3
    elif count < (split*4):
        new_dic[x] = 4
    else :
        new_dic[x] = 5
```

```python
# 단어사전 (Level1일수록 빈도수가 높은 단어)
new_dic
```

```python
#단어사전을 이용하여 각 기사별 난이도 점수화
train_score=[]
for x in train_word :
    sum = 0
    for i in x:
        sum += new_dic[i] 
    train_score.append(sum/len(x))
    
```

```python
train_df["score"] = train_score
train_df
```

```python
#빈도수 분포 확인
sns.distplot(train_df['score'])
```
